```python
import matplotlib.pyplot as plt
import numpy as np

def koch(p1, p2, n):
    if n == 0:
        return [p1, p2]
    else:
        p1_new = p1 + (p2 - p1) / 3
        p3 = p1 + (p2 - p1) / 3
        angle = np.pi / 3
        rotation = np.array([[np.cos(angle), -np.sin(angle)], [np.sin(angle), np.cos(angle)]])
        p4 = p3 + np.dot(p2 - p1, rotation) / 3
        p5 = p1 + (p2 - p1) * 2 / 3
        return koch(p1, p1_new, n - 1) + koch(p1_new, p3, n - 1) + koch(p3, p4, n - 1) + koch(p4, p5, n - 1) + koch(p5, p2, n - 1)

def draw_koch_snowflake(level):
    p1 = np.array([-200.0, 100.0])
    p2 = np.array([200.0, 100.0])
    p3 = np.array([0.0, 346.41])

    points = koch(p1, p2, level) + koch(p2, p3, level) + koch(p3, p1, level)

    x = [point[0] for point in points]
    y = [point[1] for point in points]

    plt.figure(figsize=(100, 100))
    plt.plot(x, y, 'b')
    plt.gca().set_aspect('equal', adjustable='box')
    plt.axis('off')
    plt.show()

level = int(input("Introduzca el pedido de copos de nieve de Koch: "))
draw_koch_snowflake(level)
```
